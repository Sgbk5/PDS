```python
#Reproducibility step-3 data analysis is performed on student data
#importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the student performance dataset
data = pd.read_csv("../data_clean/data_clean_student.csv")

# Set custom color palette for the visualizations 
custom_palette = sns.color_palette("Paired")  

# Define the desired order for race/ethnicity categories
race_order = ['group A', 'group B', 'group C', 'group D', 'group E']

# Visualization 1: Count plot for race/ethnicity categories based on gender count
plt.figure(figsize=(8, 6))
sns.countplot(x='race_ethnicity', data=data, hue='gender', palette=custom_palette, order=race_order)
plt.title("Count of Students by Race/Ethnicity and Gender")
plt.xlabel("Race/Ethnicity")
plt.ylabel("Count")
plt.show()
```

```python
#Visualization 2: Histogram representing the range of reading, writing and math scores.
plt.figure(figsize=(8, 6))
plt.subplot(2, 2, 1)
plt.hist(data['writing_score'], bins=10, color='pink', edgecolor='black')
plt.title('Distribution of Writing Score')
plt.xlabel('WRITING SCORE')
plt.ylabel('Frequency')
plt.xlim(0, 100)

plt.subplot(2, 2, 2)
plt.hist(data['reading_score'], bins=10, color='violet', edgecolor='black')
plt.title('Distribution of Reading Score')
plt.xlabel('READING SCORE')
plt.ylabel('Frequency')
plt.xlim(0, 100)

plt.subplot(2, 2, 3)
plt.hist(data['math_score'], bins=10, color='yellow', edgecolor='black')
plt.title('Distribution of Math Score')
plt.xlabel('MATH SCORE')
plt.ylabel('Frequency')
plt.xlim(0, 100)

plt.tight_layout(pad=2.0)
```

```python
#Visualization 3: Histogram representing and test scores based on gender.
gender_scores=data[["gender", "math_score", "reading_score", "writing_score"]]
dfm = gender_scores.melt(id_vars='gender',var_name ='score type', value_name ='mean score')
df_mean = dfm.groupby(['gender', 'score type']).mean()
custom_palette = ["red", "pink"]
ax = sns.histplot(data=df_mean, x='score type', weights='mean score', hue='gender',multiple="dodge", shrink=.7,palette=custom_palette)
plt.show()
```

```python
#Visualization 4: Create box plots for each score type, test preparation course, and gender
plt.figure(figsize=(16, 8))

plt.subplot(1, 3, 1)
sns.boxplot(x='test_preparation_course', y='math_score', data=data, hue='gender')
plt.title('Math Score vs Test Preparation Course by Gender')

plt.subplot(1, 3, 2)
sns.boxplot(x='test_preparation_course', y='writing_score', data=data, hue='gender')
plt.title('Writing Score vs Test Preparation Course by Gender')

plt.subplot(1, 3, 3)
sns.boxplot(x='test_preparation_course', y='reading_score', data=data, hue='gender')
plt.title('Reading Score vs Test Preparation Course by Gender')

plt.tight_layout()
plt.show()
```

```python
#Visualization 5: Calculate the correlation matrix with numeric columns only
correlation_matrix = data.corr(numeric_only=True)

# Create a colormap
custom_colormap = sns.color_palette("coolwarm", as_cmap=True)

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap=custom_colormap, linewidths=0.5)
plt.title('Correlation Matrix (Writing Score, Reading Score, and Math Score)')
plt.show()
```
